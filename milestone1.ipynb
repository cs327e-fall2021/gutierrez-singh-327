```python
%%timeit

import mysql.connector

connection = mysql.connector.connect(
  host="172.24.144.3",
  user="root",
  password="Renzoe22",
  database="load_testing",
  autocommit=False
)

sql = "INSERT INTO Person (first_name, last_name, company_name, address, city, county, state, zip, phone1, phone2, email, web)\
  VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"

total_counter = 0
counter = 0
try:
    cursor = connection.cursor()
    for row in rows:
        counter += 1
        total_counter += 1
        cursor.execute(sql, row)
        if counter >= 5000:
            counter = 0
            connection.commit()
            print(total_counter, "records inserted successfully into Person table")
    connection.commit()
    cursor.close()
    
except mysql.connector.Error as error:
    print("Failed to insert record into Person table {}".format(error))

finally:
    if connection.is_connected():
        connection.close()
        print("MySQL connection is closed")
    
```
